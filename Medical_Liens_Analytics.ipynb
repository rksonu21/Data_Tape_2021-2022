```python
#Import Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv
df = pd.read_csv('777 Data Tape 2021-2022 (F4D).xlsx - Sheet1.csv', low_memory = False)
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df.columns = df.iloc[0]
df.reset_index()  
```

```python
df = df.drop(df.index[0]).reset_index(drop=True)
df.head(20)  
```

```python
df.describe()
```

<b>It can be seen upon visual inspection that there exist columns that will not have any tangible contribution to the analytics process that we are about to perform. In order to optimise the execution of our algorithms and avoid wasing of memory resources, we would be better-off dropping these columns.</b>

```python
columns_to_drop = ['check','UNIQUE_ID','Plaintiff Last Name','Plaintiff First Name','Portfolio Id','Portfolio']
df = df.drop(columns = columns_to_drop)
```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'], errors='coerce')
df['Last Collection Date'] = pd.to_datetime(df['Last Collection Date'], errors='coerce')
```

```python
# Remove non-numeric characters and convert to float, handling errors
df['Billing Amount'] = df['Billing Amount'].replace({'\$': '', ',': ''}, regex=True)
df['Total Capital Advanced'] = df['Total Capital Advanced'].replace({'\$': '', ',': ''}, regex=True)
df['$_Collection'] = df['$_Collection'].replace({'\$': '', ',': ''}, regex=True)

# Use pd.to_numeric to convert to float, coercing errors to NaN
df['Billing Amount'] = pd.to_numeric(df['Billing Amount'], errors='coerce')
df['Total Capital Advanced'] = pd.to_numeric(df['Total Capital Advanced'], errors='coerce')
df['$_Collection'] = pd.to_numeric(df['$_Collection'], errors='coerce')

pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
with pd.option_context('display.float_format', '{:,.2f}'.format):
    print("\nDescription of the DataFrame:")
    print(df.describe())
```

```python
condition = (df['Lien Status'] == 'Settled') & (df['Case Status'] == 'Closed')

# Create two new DataFrames based on the condition
df_closed = df[condition]
df_open = df[~condition]
```

# Exploratory Data Analysis


Closed cases vs Open Cases

```python
# Create a count plot for 'Case Status'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Case Status', palette='viridis')
plt.title('Count of Open and Closed Cases')
plt.xlabel('Case Status')
plt.ylabel('Count')
plt.show()
```

```python
df_open['Log Billing Amount'] = np.log1p(df_open['Billing Amount'])

# Create a histogram plot for 'Billing Amount' for Open Cases
plt.figure(figsize=(10, 6))
sns.histplot(df_open['Log Billing Amount'].dropna(), bins=100, kde=True)
plt.title('Distribution of Log Billing Amount')
plt.xlabel('Log Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
df_closed['Log Billing Amount'] = np.log1p(df_closed['Billing Amount'])

# Create a histogram plot for 'Billing Amount' for Closed Cases
plt.figure(figsize=(6, 6))
sns.histplot(df_closed['Log Billing Amount'].dropna(), bins=10, kde=True)
plt.title('Distribution of Log Billing Amount')
plt.xlabel('Log Billing Amount')
plt.ylabel('Frequency')
plt.show()
```
